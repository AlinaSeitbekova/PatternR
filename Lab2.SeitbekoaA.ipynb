```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime

oper = pd.read_csv('1 - operations.csv',sep=',')
oper[:10]
```

```python
dataset = pd.read_csv('2 - dataset.csv',sep=',')
dataset[:10]
```

```python
table1 = pd.merge(oper, dataset, on='client_id')
table1[:10]
```

```python
types = pd.read_csv('3 - types.csv',sep=';')
types[:10]
```

```python
codes = pd.read_csv('4 - codes.csv',sep=';')
codes[:10]

```

```python
table2 = pd.merge(table1, types, on='type')
table2[:10]

```

```python
print("merged table: ")
finalTable = pd.merge(table2, codes, on='code')
finalTable[:10]
```

```python
finalTable['client_id'].hist(bins=100)
plt.show()
```

```python
finalTable['code'].hist(bins=100)
plt.show()
```

```python
finalTable['type'].hist(bins=100)
plt.show()
```
